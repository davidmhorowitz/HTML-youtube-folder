```python
import pandas as pd
```

```python
df1 = pd.read_csv('project1p2needsinflation.csv')
df2 = pd.read_csv('Project1p2_Inflation-by-year - Sheet1.csv')
```

```python
excesscolumns = df1.merge(df2, \
         left_on='Name', \
         right_on='Name', \
         how='left')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('max_seq_item', None)
```

```python
excesscolumns
```

```python
allcolumns = excesscolumns.drop(['Unnamed: 0', 'Mean', 'Salary'], axis=1)
```

```python
allcolumns
```

```python
allcolumns['Average Salary Raise (2018)'] = (allcolumns['Salary2'] - allcolumns['Salary1']) / allcolumns['Salary1'] * 100
```

```python
for key, value in (allcolumns['Salary2'] - allcolumns['Salary1']).items():
    if value <= 0:
        print(f"For {key}, this raise {value}.")
```

```python
allcolumns['Average Salary Raise (2018)'] = allcolumns['Average Salary Raise (2018)'].round(2)
```

```python
allcolumns['Average Salary Raise (2019)'] = ((allcolumns['Salary3'] - allcolumns['Salary2']) / allcolumns['Salary2'] * 100).round(2)
```

```python
allcolumns['Average Salary Raise (2020)'] = ((allcolumns['Salary4'] - allcolumns['Salary3']) / allcolumns['Salary3'] * 100).round(2)
```

```python
allcolumns['Average Salary Raise (2021)'] = ((allcolumns['Salary5'] - allcolumns['Salary4']) / allcolumns['Salary4'] * 100).round(2)
```

```python
allcolumns = allcolumns.rename(columns={'Inflation 2017' : 'Inflation (2018)', 'Inflation 2018' : 'Inflation (2019)', 'Inflation 2019' : 'Inflation (2020)', 'Inflation 2020' : 'Inflation (2021)', 'Inflation 2021' : 'Inflation (2022)'})
```

```python
allcolumns = allcolumns.drop(['Raise (2018)', 'Raise for teacher salary (2018)'], axis=1)
```

```python
allcolumns
```

```python
allcolumns.to_csv('raisesandinflation.csv')
```
